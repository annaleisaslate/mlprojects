```python
c = twint.Config()
Locations = ["London", "Singapore"]
Search_Terms = ["elonmusk", "btc", "eth", "bitcoin", "doge", "bnb", "xrp", "cryptocurrency"]
Start_Date = "2022-02-01"
End_Date = "2022-03-02" #this is the cutoff date

storage = list()

for i in Locations:
    for j in Search_Terms:
        c.Search = j
        c.Limit = 100
        c.Near = i
        c.Lang = "en"
        c.Since = Start_Date
        c.Until = End_Date
        c.Pandas = True
        try: 
            twint.run.Search(c)
            df_i=twint.storage.panda.Tweets_df
            storage.append(df_i)
        except:
            pass
data = pd.concat(storage, ignore_index = True)

df = data[["id","data","username","name","tweet","near"]].copy()
df.head()
```
